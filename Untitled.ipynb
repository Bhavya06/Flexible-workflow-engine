```python
import nltk
nltk.download('stopwords')
import string

#a = 'I want to make a @workflow: Student Declaration Submission. The first @Task: Task1 and @Time: 8 sec. Next @Task: Task2 and @Time: 7 sec OR @Task: Task3 and @Time: 6 sec. Third @Task: Task4 and @Time: 4 sec. Final @Task: Task5 and @Time: 3 sec. '

a = 'I want to make a @workflow: Demo. The first @Task: Task1 and @Time: 5 sec. Next @Task: Task2 and @Time: 7 sec OR @Task: Task3 and @Time: 4 sec. Final @Task: Task4 and @Time: 2 sec. '


from nltk.corpus import stopwords
stopwords = list(stopwords.words('english'))
stopwords.extend(list(string.punctuation))  
stopwords.append("i\'ve")
stopwords.append("i\'m")
stopwords.remove("no")
stopwords.remove("not")
stopwords.remove("than")
stopwords.remove("which")
stopwords.remove("or")


def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in stopwords])
b = remove_stopwords(a.lower())
b = b.replace("sec",'')
print(b)
workflow_name = ''
tasks = []
time = []
link = []
i=0
prev = []
temp =[]
or_flag = 0
or_pos = -1
while(i<len(b)):
    if(b[i] == "@"):
        ind = b[i:].index(":")
        word = b[i+1:ind+i]
        if(word == 'workflow'):
            fs = b[i+ind+1:].index('.')
            workflow_name = b[ind+2+i:i+ind+1+fs]
        elif(word=='task'):
            at = b[i+ind+1:].index('@')
            tasks.append(b[i+ind+2:at+ind+i])
            fs = b[ind+1+i:].index('.')
            if('or' in  b[ind+1+i:]):
                or_pos = b[ind+1+i:].index('or')
            else:
                or_pos = -1
            if(or_pos!=-1):
                if(fs < or_pos):
                    colon = b[ind+1+i:].index(':')
                    time.append(b[colon+i+3+ind:i+ind+fs])
                else:
                    colon = b[ind+1+i:].index(':')
                    time.append(b[colon+i+3+ind:i+ind+or_pos-1])
            else:
                colon = b[ind+1+i:].index(':')
                time.append(b[colon+i+3+ind:i+ind+fs])
            if(temp == []):
                temp = [tasks[-1]]
            else:
                if(or_pos!= -1 and or_pos < fs):
                    prev = temp.copy()
                    temp = [tasks[-1]]
                    link.append([prev[0],temp[-1]])
                    or_flag = 1
                else:
                    if(or_flag == 0):
                        prev = temp.copy()
                        temp= [tasks[-1]]
                        if(len(prev) == 1):
                            link.append([prev[0],temp[-1]])
                        else:
                            for k in range(0,len(prev)):
                                link.append([prev[k],temp[-1]])
                                print('$$$$$$',link)
                    else:
                        temp.append(tasks[-1])
                        link.append([prev[0],temp[-1]])
                        or_flag = 0
                print('******', prev)
                print("*****",temp)
        if(or_pos!=-1 and or_pos<fs):
            i=or_pos+1+ind+1+i
        else:
            i=fs+1+ind+1+i
    else:
        i=i+1
print(workflow_name)
print(tasks)
print(time)
print(link)
```
